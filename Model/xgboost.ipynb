```python
from xgboost import XGBRegressor
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.inspection import permutation_importance
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv("hourly_load_data_1.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
s1= MinMaxScaler(feature_range=(0,1))
X= s1.fit_transform(df)
```

```python
s2=MinMaxScaler(feature_range=(0,1))
y= s2.fit_transform(df[["Total_load"]])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,random_state=42,shuffle=False)
```

```python
model=XGBRegressor(tree_method='approx',subsamples=0.7,n_estimators=300,max_depth=12,learning_rate=0.5,colsample_bytree=0.6)
import time
start=time.time()
model.fit(X_train,y_train)
end=time.time()
continential= end-start
continential
```

```python
start=time.time()
pred = model.predict(X_test)
end=time.time()
continential=end-start
continential
pred
```

```python
pred.shape
```

```python
prediction = pred.reshape(pred.shape[0], 1)
```

```python
prediction.shape
```

```python
Pred= s2.inverse_transform(prediction)
```

```python
Actual= s2.inverse_transform(y_test)
```

```python
x=Pred
z= Actual
```

```python
df= pd.DataFrame(x,columns=["prediction"])
data= pd.DataFrame(z,columns=["Actual"])
```
